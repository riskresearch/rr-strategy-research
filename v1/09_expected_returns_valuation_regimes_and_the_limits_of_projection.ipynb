# 9. Expected returns, valuation regimes, and projection

At some point every investment process has to move from classification to projection. It is not enough to separate beta from alpha, or valuation from narrative, or diversification from concentration. Capital still has to be allocated, and allocation requires a view—explicit or implicit—on future return.

This is where investment work often becomes less careful. In some cases, recent returns are projected forward with only minor adjustment. In others, long-run historical averages are treated as if they were reliable guides regardless of valuation, policy, or macro regime. In others, forecasts are expressed with a degree of precision that the underlying uncertainty does not justify. None of these approaches is satisfactory. A more serious framework treats expected return as a structured estimate under uncertainty rather than as a number to be filled into a model for convenience.[^1][^2][^3]

## Return starts from the price paid

The most basic principle in return projection is that expected return depends on entry price.

This is obvious in theory and often neglected in practice. A strong asset purchased at an undemanding valuation may produce attractive long-run returns even if the path is uneven. The same asset purchased at an extreme valuation may deliver poor subsequent returns despite continued operational quality. Campbell and Shiller’s work remains one of the clearest empirical statements of this point: valuation ratios have historically contained information about long-run subsequent stock returns, mainly through future price adjustment rather than through reliable forecasting of long-run dividend growth.[^1]

In equities, the starting point for long-horizon return usually contains at least three elements: the cash yield available to shareholders, the growth of underlying earnings or cash generation, and the change in the valuation multiple applied by the market. Research Affiliates summarizes the arithmetic succinctly: expected equity return can be decomposed into starting yield, real growth, inflation, and valuation change.[^2] This decomposition is simple, but it remains useful because it prevents return projection from becoming a vague expression of optimism or caution.

If a market is trading at a low yield and a high multiple, strong future returns may still occur, but they require more to go right. Growth must remain strong, margins must hold, and valuation must stay elevated or expand further. If a market begins at a lower valuation and a higher yield, the future does not need to be as generous for returns to be acceptable. The point is not that valuation determines short-run timing perfectly. It is that valuation enters the arithmetic of long-run return whether investors prefer to speak about it or not.[^1][^2]

The same logic applies to bonds, though in a different form. Starting yield is often a strong anchor for future return if the bond is held over a sufficient horizon and default is not a central concern. Antti Ilmanen’s survey of expected returns across major asset classes makes this point directly: for high-quality bonds, beginning yield is usually the most informative starting estimate for medium- to long-horizon returns, even though mark-to-market outcomes can vary materially along the way.[^3] Recent Vanguard work makes a similar practical argument, emphasizing that higher starting yields materially improve the resilience of future bond returns even when yields rise modestly further.[^4]

## Long-run averages are useful and dangerous

Historical return data is indispensable, but it is also easy to misuse.

Long-run averages provide perspective. They show what different asset classes have delivered through multiple cycles, inflation regimes, and policy environments. They remind investors that equities have generally offered higher returns than cash and high-quality bonds, and that some strategies earn their keep by carrying unpleasant exposure through stress. But long-run averages become dangerous when they are used without regard to starting conditions.[^3][^5]

A historical equity return observed across many decades includes periods that began at low valuations, high valuations, stable inflation, unstable inflation, easy policy, tight policy, rapid productivity growth, war, recession, and disinflation. To lift that average out of history and treat it as a clean forecast for the next decade is too careless. Campbell and Shiller are explicit on this point: valuation ratios are more useful for thinking about long-run expected returns than for assuming that the unconditional historical average is always the right forecast.[^1]

The same issue applies to bonds. A long period of falling yields can produce realized returns that should not be expected to repeat once yields are already low. Conversely, a period of rising yields can make the asset class look unattractive just when future carry has improved materially. History is most useful when it is used to frame possibility, range, and regime behavior. It is less useful when it is used as a substitute for thinking about the present starting point.[^3][^4]

## Regimes matter because pricing is not static

Expected returns are not produced in a neutral environment. They are produced within regimes.

A regime is not simply a market trend. It is a broader configuration of growth, inflation, liquidity, policy, financing conditions, and investor tolerance for risk. Different regimes change both the level of return available and the type of return that is likely to dominate. Ilmanen’s framework is particularly useful here because it treats expected returns as time-varying and shaped by both valuation and macro conditions rather than as stable long-run constants.[^3][^5]

In a falling-rate, low-inflation environment, long-duration assets may benefit from both fundamental growth and valuation expansion. In a rising-rate or inflationary environment, those same assets may face compression even if the underlying businesses remain sound. In a liquidity-rich environment, credit spread, carry, and valuation extension may all look more stable than they really are. In a stressed or tightening regime, fragility appears quickly and projected return has to be revised downward.[^4][^5]

This is one reason expected return should not be modeled as if macro conditions were merely background noise. The macro regime does not determine every outcome, but it changes the discount rate applied to future cash flows, the tolerance for leverage, the willingness to pay for duration, and the reliability of diversification relationships. All of these matter directly for what an investor should reasonably expect to earn.

## Projection should be done in ranges, not single-point confidence

One of the easiest ways to make expected returns look more knowable than they are is to collapse them into a single number.

That number may be convenient for optimization or reporting, but it often creates false precision. Expected return is better treated as a range tied to explicit assumptions. The range may not be wide enough to satisfy philosophical uncertainty, but it should be wide enough to reflect real sensitivity to growth, inflation, margins, discount rates, default conditions, and valuation changes. Ilmanen and Research Affiliates both argue for this more conditional style of forecasting rather than for overconfident point estimates.[^2][^3]

This does two useful things. First, it makes disagreement more productive. Two investors may not differ in principle; they may differ only on one or two assumptions that can be identified clearly. Second, it improves portfolio design because it reduces the temptation to size aggressively off a false sense of certainty.

A useful projection framework should therefore distinguish between base case, favorable case, and adverse case. It should also indicate which variable is doing most of the work. If the expected return only looks attractive when multiple expansion continues, that should be stated plainly. If the return depends mainly on carry rather than on price appreciation, that should also be stated plainly.

## Stocks and bonds should not be projected by the same logic

It is common to discuss expected returns from stocks and bonds side by side, but they should not be forecast with identical reasoning.

Equities are claims on uncertain future cash generation. Their long-run return reflects cash yield, growth, dilution or buyback behavior, and changes in the multiple the market assigns. The range of outcomes is wide because both the cash flows and the multiple are unstable.[^1][^2]

Bonds are more bounded, but not trivial. Yield provides a starting anchor, especially for high-quality fixed-income instruments held over time. But duration risk, inflation, real-rate changes, credit deterioration, and reinvestment conditions still matter greatly. A bond with a respectable starting yield may produce disappointing real return in a persistent inflation regime. A long-duration bond may suffer a deep mark-to-market drawdown before its contractual cash flows reassert themselves. A credit instrument may appear attractive on spread until liquidity stress or refinancing difficulty reveals how much of the return was compensation for fragility.[^3][^4]

This is why bonds should not be described simply as safe and equities as risky. Each has its own return logic, and each can disappoint in ways that are specific to the structure of the claim.

## The main forecasting error is usually overstating stability

Most errors in expected return work do not come from arithmetic. They come from assuming more stability than the environment justifies.

Margins are assumed to mean-revert gently when in reality they may be more cyclical. Growth is assumed to normalize when in reality competition or regulation may alter the path more sharply. Policy is assumed to remain supportive when in reality inflation or financial instability forces a change. Correlations are assumed to help because they recently have done so. A strategy with smooth realized history is assumed to remain smooth because its hidden dependence has not yet been tested.[^3][^5]

This is the point at which return projection becomes most dangerous. The investor is no longer simply estimating a future path. The investor is importing a view of structural stability without stating it clearly. Once that hidden assumption is recognized, the need for scenario work becomes much clearer.

This is also where humility becomes practical rather than philosophical. Expected return work improves less through increasingly elaborate models than through a cleaner understanding of what assumptions are carrying the forecast.

## What this means for investment practice

The practical lesson is not to stop making return forecasts. Capital allocation requires them. The lesson is to make them in a way that is more conditional, more transparent, and less dependent on recent experience.

That means beginning from valuation and carry rather than recent performance alone. It means adjusting the forecast for macro regime and financing conditions rather than assuming those variables will remain benign. It means distinguishing between return that depends on multiple expansion and return that can be earned with little help from repricing. It means using ranges and scenarios rather than pretending the future can be reduced to one clean number.[^2][^3][^4]

Most importantly, it means treating expected return as one part of a broader portfolio decision rather than as a stand-alone truth. A moderate expected return with strong diversification value may deserve more capital than a higher expected return that is heavily dependent on one crowded regime. The forecast has to be judged together with path, correlation, liquidity, and implementation reality.

## Closing remark

Expected return is central to investment work precisely because it cannot be known with confidence.

That is not a reason to avoid the exercise. It is a reason to handle it with more care. The investor who begins from entry price, respects the importance of regime, works with ranges rather than false precision, and understands the different logic of stocks and bonds is not solving uncertainty. But that investor is at least framing it properly.

That is the standard that matters. Return projection should not pretend to remove ambiguity. It should make the structure of that ambiguity more visible so that capital can be allocated with better judgment and fewer hidden assumptions.
