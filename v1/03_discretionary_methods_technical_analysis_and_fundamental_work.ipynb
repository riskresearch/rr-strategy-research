# 3. Discretionary methods, predictability, and the limits of market signals

Once price, value, and crowd behavior have been separated conceptually, the next question is practical: how do investors actually try to form an edge?

In broad terms, most discretionary investment work falls into two families. One begins from the asset itself and tries to estimate its worth through business quality, cash flows, industry structure, balance-sheet strength, and valuation. The other begins from the market and tries to infer future price movement from price action, volume, trend, relative performance, and related trading information. These two traditions are usually called fundamental analysis and technical analysis. The distinction is useful, but it should not be overstated. In practice many investors mix them, whether they admit it or not.[^1][^2]

This matters because both traditions are often presented badly. Technical analysis is frequently described either as useless pattern worship or as a complete trading solution. Fundamental analysis is often described either as the only serious approach or as a theoretically elegant but practically slow exercise with little timing power. The more productive task is to understand what each method is actually trying to do, where each can add value, and where each tends to fail.[^2][^3]

## Fundamental work starts from the asset

Fundamental analysis starts from the view that the asset has economic characteristics that matter for long-run return. In equities, this usually means revenue quality, margins, capital intensity, balance-sheet strength, industry structure, management incentives, reinvestment opportunities, and valuation. In credit, it means debt capacity, covenant structure, refinancing exposure, collateral, and default risk. In macro and cross-asset work, the equivalent may involve growth, inflation, policy, fiscal conditions, external balances, and the sensitivity of markets to those variables. That is the central appeal of fundamental work: it begins with the economics of the object being owned rather than with the path of the quotation alone.[^4]

At its best, fundamental work is trying to answer a serious question: what is being owned, what is likely to happen to it economically, and what would justify owning it at the current price? That remains essential. It forces the investor to think in terms of mechanism rather than pure pattern recognition. It also provides an anchor against which market enthusiasm or pessimism can be judged. A weak business does not become attractive simply because price action is strong. A fragile credit does not become safe because spreads have tightened.

At the same time, fundamental work has limits that are often understated by those who prefer it. Good asset-level analysis does not guarantee good returns over a relevant horizon. A stock can be well understood and still be a poor investment if expectations are already rich, if the market is focused on a different risk, or if there is no reason for capital to reprice the asset in the investor’s timeframe. This is one reason good company analysis and good investment results are not the same thing.[^1][^4]

## Price-based work starts from the market

Technical analysis begins from a different premise. Instead of asking first what the asset should be worth, it asks what the market itself may be revealing through price and trading behavior.

At its weakest, this becomes a catalogue of patterns, lines, and formations with little stability or explanatory value. That version deserves the criticism it often receives. But that is not the whole field. At its more serious end, price-based work is an attempt to study how supply, demand, trend persistence, reversals, and relative strength express themselves in market prices. It treats price as information rather than as a mere output.[^2][^3]

That is more defensible than many fundamental investors admit. Price can contain information that a purely asset-centered process misses. It can reflect changing expectations before those expectations appear in reported numbers. It can reveal crowding, forced positioning, narrowing or broadening participation, and shifts in risk appetite. Academic work that tests simple trading rules and return continuation does in fact find structure in historical prices, even if the interpretation and persistence of that structure remain contested.[^3][^5][^6]

A simple example is a security that begins to underperform persistently despite appearing optically cheap and despite no major deterioration in the headline narrative. That price behavior does not prove the market is correct, but it does suggest that some combination of expectations, positioning, or information is moving against the thesis. Ignoring that signal entirely in the name of analytical purity is often just another way of ignoring adverse evidence.

Still, price-based work has serious limits. Price can reveal behavior, but not always cause. A trend may persist for fundamental reasons, for flow reasons, or for reasons that cannot be cleanly separated in real time. A reversal pattern may matter in one regime and fail completely in another. A rule that backtests well may decay once it is crowded, arbitraged, or exposed to a different market structure. The weakness of technical work is not that it uses price. The weakness appears when price is treated as self-sufficient explanation.[^2][^3]

## Predictability is limited, but not absent

Once these two methods are set side by side, the central issue becomes clearer: how much predictability do markets actually allow?

This question is often handled badly. One side treats markets as so efficient that almost all attempts at prediction are futile. The other points to recurring patterns, anomalies, and episodes of obvious excess as proof that prices are broadly forecastable. Neither position is satisfactory. Markets are not so efficient that price behavior becomes meaningless, but neither are they so predictable that abnormal return can be extracted easily and indefinitely once a pattern becomes widely known.[^1][^7]

A more serious approach avoids absolutes. The useful question is not whether markets are efficient in a doctrinal sense. The useful question is what kind of predictability exists, at what horizon, under what conditions, and whether that predictability survives implementation, crowding, and regime change. Fama’s later review is useful on this point: return predictability and market efficiency are not clean opposites, because predictability may reflect time-varying expected returns rather than simple exploitable error.[^7]

In practical terms, market efficiency does not mean that prices are always correct. Nor does it mean that all market participants are rational, that bubbles cannot occur, or that prices move only in response to perfectly processed information. The more useful meaning is narrower: easily observable opportunities for abnormal risk-adjusted return are difficult to exploit persistently because competition acts on them quickly.[^1][^7]

This distinction matters because many criticisms of market efficiency attack a weaker version of the idea than serious investors actually care about. The existence of momentum, reversal, carry, or seasonal effects does not automatically prove that returns are broadly forecastable in a way that remains investable after costs and crowding. But the evidence also does not support the strongest version of the random-walk view. Lo and MacKinlay reject a simple random walk for stock returns, and Brock, Lakonishok, and LeBaron find that simple technical trading rules contain information in historical data. Jegadeesh and Titman show medium-horizon momentum in stock returns. These findings do not settle the debate, but they do show that price data has sometimes contained structure beyond pure white noise.[^3][^5][^6][^8]

## What market signals are really trying to capture

When market-based methods claim predictive power, they are usually trying to capture one of a small number of mechanisms.

The first is trend persistence. Markets sometimes continue moving in one direction because information diffuses gradually, because positioning adjusts over time rather than instantly, or because institutional flows reinforce an existing move. Momentum research is the clearest example of this family of effects.[^6][^9]

The second is reversal. Markets also overshoot. When positioning becomes crowded, when forced buying or selling exhausts itself, or when expectations move too far relative to what can reasonably occur, price can reverse sharply. Some of the early overreaction literature and later contrarian evidence fit this logic, though the horizon and implementation details matter a great deal.[^10]

The third is relative strength. Here the signal is not simply that a market has risen or fallen, but that it is behaving differently from peers, benchmarks, or related assets. Sometimes this reflects changing expectations or a regime shift before that shift is fully visible elsewhere.

The fourth is participation and breadth. A move carried by a narrowing group of assets or by shallow participation may be weaker than it appears. Conversely, a move that broadens through the cross-section may reveal more durable underlying support. In practice, many discretionary and systematic investors use this information even when they do not describe it as formal technical analysis.

These are defensible objects of study. The problem arises when pattern language becomes detached from mechanism. A chart formation with no explanation beyond visual familiarity is a weak basis for investment. A recurring historical feature that can plausibly be tied to slow information diffusion, behavioral underreaction, position adjustment, or flow persistence is a much stronger candidate for research.[^3][^6][^9]

## Why many signals disappoint

The fact that a signal exists historically does not tell us much on its own. Most signals disappoint for one of three reasons.

The first is that they were never economically meaningful to begin with. They appear significant in the sample but have no durable mechanism behind them. Once conditions change, they disappear.

The second is that they were meaningful but became crowded. This is especially common in liquid markets. A signal that worked when it was obscure may weaken once capital systematically targets it, reducing forward returns or making the path less attractive.

The third is that the signal depends on regime. Some behaviors survive only when inflation is stable, when policy is predictable, when liquidity is abundant, or when dispersion across assets is high enough for relative signals to matter. A strategy can therefore fail not because the original idea was foolish, but because the environment in which it worked has changed. Lo’s adaptive-markets perspective is useful here: market efficiency and predictability can vary with institutional conditions and competition rather than remaining fixed states.[^2]

This is one reason investment work should be careful with historical certainty. A pattern that has worked repeatedly in the past is a useful starting point, but not a final answer. The research task is to ask why it worked, why it may continue to work, and what would cause that logic to weaken.

Randomness and structure can coexist. A market can be noisy at short horizons and still exhibit persistent structure at longer ones. It can be difficult to forecast day to day and still reward certain forms of trend, carry, relative value, or valuation over time. It can also display visible inefficiencies while remaining extremely hard to beat in practice because implementation is costly, timing is uncertain, or the pattern is fragile. That balance is where serious investment work should stand.[^2][^7]

## Closing remark

The most useful standard is not whether a method is called fundamental or technical. The more useful standard is whether it extracts information that is economically or behaviorally meaningful, whether it can be implemented with restraint, and whether it survives changing market conditions.

Fundamental work asks what the asset is worth and why. Price-based work asks what the market is doing and what that behavior may reveal. Each contains useful information. Each also contains ways to mislead. Market efficiency does not eliminate the relevance of either approach. It simply raises the standard required before a signal deserves capital.

The task is not to choose a camp for reasons of identity. The task is to understand which information is being produced, which horizon it belongs to, and whether it helps explain future return rather than merely describe the past. That is the standard that matters for the rest of this framework.
