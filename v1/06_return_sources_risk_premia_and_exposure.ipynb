# 6. Return sources, risk premia, and exposure

Investment returns are often discussed as if they were one thing measured on one scale. In reality they are generated through different mechanisms.

A broad equity index, a long-duration government bond, a credit portfolio, a carry strategy, and a concentrated stock portfolio may all produce attractive returns over long periods, but they do so for different reasons. Some are exposed to growth, some to inflation, some to liquidity conditions, some to term structure, some to behavioral persistence, and some to the willingness of investors to bear drawdown and uncertainty. Asset-pricing research formalizes this by decomposing returns into exposures to common drivers rather than treating all return as one homogeneous object.[^2][^3]

This distinction matters because not all returns should be judged in the same way. A long-only equity return should not be interpreted in the same manner as a market-neutral relative value return. A strategy with stable carry and occasional crash risk should not be treated as if it were equivalent to a strategy with low carry and high convexity. A return stream produced mainly through passive market exposure should not be described as manager skill simply because it was held through a favorable period.[^1][^3][^5]

The point is not to dismiss high return where it appears. The point is to understand what generated it, what risks accompanied it, and whether those risks were consciously chosen or merely inherited.

## Risk premia are compensation, not free return

The idea of a risk premium is straightforward in principle. Investors are compensated for holding exposures that are uncertain, economically sensitive, illiquid, cyclically painful, or difficult to own when stress is concentrated. Equity risk is the most familiar example. Long-run equity returns have historically exceeded those of cash because equities expose the holder to business risk, recession risk, valuation compression, and significant drawdown. Standard asset-pricing theory begins exactly from this logic: higher expected return is compensation for bearing systematic risk that cannot be diversified away.[^2][^6]

That logic extends beyond equities. Duration, credit, carry, and certain forms of volatility selling can all be understood partly as compensation for bearing a type of risk that others are unwilling or unable to hold without payment. In credit, spread compensates investors for expected losses, default uncertainty, liquidity stress, and cyclical vulnerability. In currency markets, carry returns have repeatedly been linked to exposure to global volatility and crash risk rather than to a simple free lunch.[^7][^8]

The practical importance of this idea is that not all positive return reflects superior forecasting. Some return simply reflects that the investor was paid to bear discomfort. This does not make the return less valuable. It makes it more intelligible. A strategy that earns a premium for carrying recession-sensitive assets should be expected to suffer when recession risk rises. A strategy that earns a premium for selling insurance-like exposure should be expected to perform poorly in sharp dislocations. If these outcomes are treated as anomalies rather than as part of the contract, the analysis has failed to identify what was being earned in the first place.[^6][^7][^8]

A useful example is credit. A portfolio of lower-quality corporate bonds may generate an attractive spread over governments for years, and that spread may look stable enough to encourage complacency. But the excess return is not mysterious. It is payment for default risk, downgrade risk, liquidity stress, and cyclical vulnerability. The danger begins when that spread is interpreted as manager alpha rather than as compensation for bearing an exposure that can deteriorate quickly under stress.[^7]

## Higher return can be disguised exposure

One of the most common errors in investment analysis is to mistake amplified exposure for skill.

This happens when a portfolio with more leverage, more duration, more credit risk, more growth sensitivity, or more illiquidity produces higher returns than a benchmark and the excess is attributed to superior judgment without properly adjusting for the underlying exposure. The manager may indeed have skill, but the return itself is not enough to prove it. This is exactly why factor models became so important: they provide a way to separate broad market exposure from systematic tilts and from residual return.[^3][^9]

A simple example would be a strategy that outperforms in a falling-rate, high-liquidity environment because it holds long-duration growth assets. The realized return may look impressive, but a large part of it may be the product of one macro exposure rather than repeated stock-picking skill. If the regime shifts and discount rates rise, the same portfolio may begin to underperform sharply. The earlier return was real, but the explanation of it was incomplete.

The same issue appears in less obvious form in multi-asset and alternatives portfolios. Some approaches look diversified on the surface while relying heavily on one hidden assumption: stable financing, benign liquidity, or the continued ability to monetize carry. Once that assumption breaks, the return stream often changes character quickly.

## Not all excess return is a premium

Although many returns are compensation for bearing risk, some returns do appear to arise from the way markets are structured and from the way investors behave within them.

Momentum is one example often discussed this way. Trend persistence and relative strength may reflect slow information diffusion, institutional delay, benchmark effects, or the tendency of investors to underreact initially and then chase later. Carry in some markets may partly reflect hedging pressure or balance-sheet segmentation rather than only classical macro risk. Certain relative value opportunities may persist because the capital required to exploit them is limited, constrained, or uncomfortable to deploy at scale. The academic literature does not treat these effects as identical, but it does show that return regularities such as momentum, value, carry, and quality have historically appeared across long samples and multiple markets.[^3][^8][^10][^11]

This is where investment research has to stay careful. It is too easy either to romanticize these return streams as pure alpha or to dismiss them as nothing more than hidden beta. In practice, many of them are mixed. They involve some form of exposure that is costly to bear and some form of market structure that allows the return to persist longer than a fully frictionless model would suggest.

Quality is a useful example. High-quality firms—those with stronger profitability, stability, and balance-sheet characteristics—have historically delivered returns that appear stronger than their valuations alone would imply, a pattern Asness, Frazzini, and Pedersen formalize in their "quality minus junk" framework. Whether one interprets this as a neglected characteristic, a behavioral anomaly, or a form of mismeasured risk, it is not well described by market beta alone.[^11]

## Beta, factor exposure, and alpha should be separated more carefully

The language of active management becomes clearer once return is divided into three broad components.

The first is broad market exposure, often captured under the language of beta. This is the return associated with participating in a major market or broad factor complex. It is not manager skill, though it may still be valuable. The classical CAPM gives the simplest version of this logic by tying expected return to market beta.[^2][^6]

The second is systematic factor exposure. This sits somewhere between broad market beta and pure alpha. It includes things like value, momentum, carry, quality, duration, or other persistent return drivers that can often be identified across time and assets. These exposures may reflect risk premia, structural segmentation, or behavioral effects. Their presence does not eliminate the possibility of skill, but it changes how performance should be interpreted. Fama and French’s work on common risk factors and subsequent factor research is the standard reference point here.[^3][^9]

The third is alpha in the stricter sense: return that remains after common market and factor exposures have been accounted for, and that cannot easily be explained by leverage or passive loading on known drivers. True alpha is difficult because it requires some combination of differentiated insight, implementation quality, and persistence under competition. Even then, what looks like alpha can later turn out to be exposure to a factor that was not previously measured well.[^1][^3][^9]

This separation is never perfect. Models are incomplete, factors are unstable, and residual return may still contain hidden exposures that have not yet been identified properly. But the exercise remains useful because it reduces the tendency to credit all positive return to skill.

## The price of exposure is usually paid in drawdown

A practical way to identify whether a return stream is genuinely earning a premium is to ask what pain must be endured in order to harvest it.

That pain may appear as drawdown, illiquidity, mark-to-market volatility, correlation to recession risk, sharp left-tail events, or long stretches of weak relative performance. If a return stream claims to be both structurally attractive and nearly painless, skepticism is warranted. In most cases, the payment arrives because the exposure is difficult to hold when conditions become unfavorable. This is a recurring theme in research on carry, credit, volatility selling, and even equity risk itself.[^6][^7][^8]

This is one reason a portfolio should never be judged by average return alone. The relevant issue is the distribution through which the return is earned. Some strategies bleed slowly and crash rarely. Others earn little most of the time and then perform powerfully in stress. Others collect stable carry for years and give back a large share in one compressed episode. These are different contracts with the investor, even when long-run average returns look similar.

For research purposes, this means that expected return and expected pain should be described together. A premium that cannot be held through its natural drawdown is of limited practical value, however attractive it appears in a long-run chart.

## What this means for portfolio design

Once return sources are classified more carefully, portfolio construction improves immediately.

Broad market exposure can be held deliberately rather than smuggled in under the language of skill. Factor exposures can be evaluated for overlap rather than accumulated unknowingly. True diversifiers can be recognized more clearly because the portfolio is no longer being judged only by nominal strategy labels. Most importantly, leverage can be applied with better awareness of what is actually being levered.[^3][^5][^9]

This also helps with expectation setting. A portfolio built mainly from risk premia should not be judged as if it were designed to produce smooth market-independent alpha. A portfolio that relies on concentrated fundamental judgment should not be excused by factor language when its exposures are mostly systematic. A macro portfolio driven by policy and rate sensitivity should not be evaluated as if its returns were idiosyncratic to manager insight alone.

Clearer classification does not eliminate uncertainty, but it reduces category error, and category error is one of the most avoidable weaknesses in professional investment work.

## Closing remark

Return is not self-explanatory. Before it can be judged, it has to be decomposed.

Some returns are payment for bearing risk. Some are expressions of broad market or factor exposure. Some arise from structural features of markets and investor behavior. A smaller subset may represent genuine alpha in the stricter sense. All of them can be useful in the right place, but they should not be confused with one another.

A serious research framework should therefore ask not only whether a return stream has been strong, but what kind of return stream it is, what pain is attached to it, and whether that pain has been understood properly in advance. That is the standard that makes later discussion of behavior, portfolio methods, and expected returns more grounded and more useful.
