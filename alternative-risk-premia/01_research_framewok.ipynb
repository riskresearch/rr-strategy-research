## 1. Introduction and market view

This repository begins from a simple premise: investment ideas should be judged under demanding standards because financial markets are noisy, competitive, and easy to misread. The goal is not to defend a particular doctrine or to claim that markets are always efficient. It is to establish a practical framework for interpreting evidence, comparing strategies, and distinguishing genuine value added from luck, narrative, or repackaged exposure.

The working stance is skeptical, but not dogmatic. I do not assume that prices are always correct, and I do not deny that bubbles, dislocations, and episodes of mispricing occur. I assume something narrower and more useful: liquid markets are competitive enough that persistent abnormal return should never be taken for granted. A market does not need to be perfectly efficient to be difficult to beat. It only needs to be efficient enough that any apparent edge must survive scrutiny, costs, and time.

A useful empirical reason for beginning from this skeptical stance is the evidence summarized in the SPIVA scorecards published by S&P Dow Jones Indices. SPIVA compares active funds with relevant benchmark indices using a framework that explicitly addresses survivorship bias, benchmark selection, style consistency, and both equal- and asset-weighted averages. In the U.S. large-cap category, SPIVA reports that over the 15-year period ending December 31, 2025, 89.93% of funds underperformed the S&P 500, leaving only 10.07% ahead. This does not prove that markets are perfectly efficient or that active skill does not exist. It does support a narrower and more practical conclusion: persistent outperformance is uncommon enough that it should never be presumed.

This starting point matters because research quality depends heavily on the assumptions made at the beginning. If one begins from the belief that markets are easy to forecast, almost any backtest can be made to look persuasive. If one begins instead from the belief that prices are adaptive, noisy, and heavily contested, then the burden of proof rises immediately. A strong result may reflect real structure, but it may also reflect leverage, hidden factor exposure, favorable sample selection, or chance.

That assumption has several practical consequences. First, broad passive exposure remains the natural reference point for most active decisions. A strategy should not be judged only by whether it produced a positive return, but by whether it improved meaningfully on a simple and investable alternative. Second, portfolio construction matters as much as, and often more than, isolated prediction. A weak signal may still be useful if it improves the total portfolio, while a strong standalone backtest may add little if it merely repackages familiar risk. Third, implementation belongs inside the analysis. Costs, turnover, taxes, leverage, liquidity, and benchmark choice are not minor details added at the end. They are part of the economic meaning of the result.

A further implication is that past price action alone is usually a weak basis for strong directional claims. Historical patterns can contain information, but they are not laws. A pattern may appear stable in sample and still disappear once market conditions change or once capital crowds into it. This is the practical value of the random-walk perspective: not as an absolute claim that prices are unknowable, but as a disciplined null hypothesis against easy predictability.

It is equally important to distinguish market inefficiency from tradable opportunity. Prices can overshoot, undershoot, and move to obvious extremes without offering a straightforward path to excess return. A mispricing seen clearly in hindsight may still be difficult to monetize in real time because the timing of convergence is uncertain, interim losses may be severe, leverage may be required, and trading frictions may absorb much of the edge. The relevant question is not simply whether markets are sometimes wrong, but whether a repeatable process can identify such conditions in advance and extract value from them under realistic conditions.

The same caution applies to forecasting more broadly. In financial markets, expected returns are often small relative to realized volatility. The signal is weak compared with the noise through which it must be detected. At the same time, the environment is not fixed. Liquidity, regulation, participation, correlations, and policy conditions change through time. A model that appears convincing in one regime may lose force in another. Explanation is therefore much easier than prediction. Once the path of returns is known, a coherent story can almost always be written after the fact. Real-time forecasting has no such advantage.

For that reason, the portfolio is usually a more meaningful object of study than the isolated trade. If forecasting is difficult and individual signals are fragile, then long-run results depend heavily on how exposures are combined, scaled, and rebalanced. This shifts the focus of the repository away from isolated claims of predictive power and toward disciplined comparison: compared with what benchmark, with what exposures, under what assumptions, at what cost, and with what contribution to the total portfolio?

The broader aim of this framework is modest but important. It is not to eliminate uncertainty, because that is impossible. It is to handle uncertainty more honestly by forcing claims to justify themselves under realistic standards. If the framework is useful, it will be useful for that reason: not because it promises easy answers, but because it raises the quality of the questions being asked.

## 2. Market interpretation: technical analysis, fundamental analysis, and evidence

The first major divide in investment practice is between technical and fundamental analysis. Technical analysis tries to infer future price movement from past prices, trading ranges, chart formations, momentum, volume, or related market data. Fundamental analysis tries to estimate value from earnings, cash flows, balance-sheet structure, growth prospects, competitive position, or macroeconomic context.

The distinction matters because it reflects two different views of what market information means and how it should be used.

Technical analysis is attractive because prices do contain information. They summarize supply and demand, reflect positioning, and can display persistence over some horizons. Momentum, for example, has a serious empirical history and cannot be dismissed casually. Yet that does not imply that chart patterns or visually persuasive setups automatically produce robust abnormal returns. The practical problem is that many price-based regularities are weak, unstable, or easy to overfit. Once enough people search the same historical record, spurious patterns inevitably appear.

Technical evidence therefore requires more than a convincing chart. It must survive:

- parameter changes,
- transaction costs,
- out-of-sample evaluation,
- benchmark-relative comparison,
- and plausible economic interpretation.

Fundamental analysis is often treated as more serious because it links price to economic value rather than to pattern recognition. That advantage is real, but it is not sufficient. The practical difficulty is that value itself is estimated under uncertainty. Future cash flows are unknown, discount rates change, accounting is imperfect, competition reshapes industries, and analysts are subject to both incentives and error.

Even when the conceptual framework is sensible, the output may still be too imprecise to support reliable outperformance. When fundamental analysis fails, the problem is usually not theoretical. It is empirical. The information may already be in price, the forecast may be wrong, or the path from valuation insight to realized return may be too unstable.

The important conclusion is not that one method must replace the other. It is that both technical and fundamental work should be judged by the same evidentiary standard. An investment claim should not be accepted because it sounds economically reasonable, because it matches a persuasive narrative, or because it produced a strong historical result under one convenient specification. It should be accepted only when the result is difficult to explain away as luck, hidden exposure, construction error, or selective reporting.

A practical reinforcement of that standard comes from the design of the SPIVA scorecards. Their value is not simply that they produce discouraging statistics for active managers. Their value is methodological: they make benchmark-relative comparison, survivorship treatment, and style consistency central to interpretation. In that sense, SPIVA is useful not as a final answer to every investment question, but as an example of the kind of evidentiary discipline that strategy research should aim to match.

In this framework, evidence means at least the following:

- A favorable result is not yet evidence.
- Evidence must be benchmark-relative.
- A result should persist across different sample windows.
- The conclusion should survive reasonable changes in construction.
- The mechanism should be economically interpretable.
- The result should remain meaningful after implementation costs.

These standards are demanding by design. Financial data are noisy, and a large enough search over strategies, factors, thresholds, and lookback windows will always produce attractive historical results by chance alone. Evidence begins only when the conclusion remains persuasive after the easy explanations have been removed.

## 3. Risk, return, and asset pricing

Return without risk has little analytical value. Any investment process can be made to look attractive if attention is restricted to the endpoint of the equity curve and the path taken to reach it is ignored. The central problem in portfolio research is therefore not simply how much a strategy earns, but what it must bear, absorb, and survive in order to earn it.

Two portfolios may arrive at similar long-run returns while exposing capital to very different distributions of outcomes. One may compound steadily with manageable drawdowns. The other may depend on hidden leverage, severe interim losses, or very long recovery periods. These are not equivalent results.

In formal portfolio theory, risk first enters as dispersion. If future returns are uncertain, then the spread of possible outcomes matters alongside their average level. This is the logic behind variance and volatility in Harry Markowitz’s framework: they provide a first language for comparing uncertain return streams and for understanding how diversification operates through covariance. The central insight is that assets should not be judged one by one, but by the way they contribute to the risk and return of the total portfolio. A portfolio is not a bag of securities. It is a structure of interacting exposures.

Variance, however, is only a starting point. Investors do not experience upside and downside deviations symmetrically. A gain above expectation is not felt in the same way as a loss below expectation, and a return stream with moderate volatility may still be unattractive if it contains deep drawdowns or very long recovery periods. This is why downside risk and drawdown belong beside volatility in any serious evaluation. Two strategies can have similar annualized volatility and very different investment quality if one suffers short, manageable setbacks while the other spends years below its previous peak.

A second distinction matters just as much: not all risk is rewarded in the same way. Finance separates diversifiable risk from systematic risk. Diversifiable risk is specific to individual securities or narrow exposures and can be reduced through diversification. Systematic risk is tied to broad market movements and cannot be eliminated simply by holding more names. This is the logic behind beta and the capital asset pricing model developed by William Sharpe, John Lintner, and Jan Mossin.

CAPM is useful because asks a basic but essential question: did the strategy outperform because it produced true alpha, or simply because it carried more market risk than the benchmark?

A simple numerical example makes the point. Suppose the equity market returns 10%, the risk-free rate is 2%, and a portfolio has a market beta of 1.2. Under CAPM, the expected return would be:

$$
E(R_i) = R_f + \beta_i (E(R_m) - R_f)
$$

So:

$$
E(R_i) = 2\% + 1.2 \times (10\% - 2\%) = 11.6\%
$$

If that portfolio earns 12%, the excess over its CAPM-implied return is only 0.4%. That is very different from saying it “beat the market” by 2 percentage points. CAPM removes part of the ambiguity by forcing performance to be judged relative to systematic exposure.

This is why CAPM is valuable. It turns vague claims into decomposable claims.

But it is not a complete description of expected return.

The limitation, recognized early in practice and theory, is that market beta alone does not explain all persistent differences in returns. Stephen Ross’s arbitrage pricing theory generalizes the problem by allowing expected return to depend on exposure to several systematic influences rather than on market beta alone. Eugene Fama and Kenneth French later formalized one influential empirical extension through the three-factor model, and later work expanded the set further to include profitability and investment. Momentum, associated prominently with Narasimhan Jegadeesh and Sheridan Titman in the empirical literature, adds another dimension that has often mattered in practice.

The broader point is simple: a return stream may be driven by several common sources of variation, not just by the market.

Common candidates include:

- market beta,
- value exposure,
- size exposure,
- profitability,
- investment intensity,
- momentum,
- duration,
- carry,
- credit sensitivity,
- and related factor tilts.

These models are not just statistical conveniences. They are tools for classification.

That classification role matters because many strategies that appear original are not. A process may look sophisticated in isolation and yet be doing little more than loading on equity beta, small-cap risk, value, duration, or another broad source of return. Complexity is not the same thing as distinctiveness.

A practical example helps. Imagine three equity funds over a long sample:

- **Fund A** returns 10.8% per year.
- **Fund B** returns 10.4% per year.
- **Market index** returns 10.0% per year.

At first glance, both funds appear to beat the market. But after factor decomposition:

- Fund A turns out to have a beta of 1.15 and a strong small-cap tilt.
- Fund B has beta near 1.00, but a persistent value loading.

If Fund A’s extra return vanishes after adjusting for higher market and size exposure, then what looked like alpha was mostly compensated exposure. If Fund B’s return vanishes after controlling for value, then the manager may have delivered value exposure in a more expensive wrapper. In both cases, the question is not whether the manager was active. The question is whether the activity added anything that could not have been obtained more cheaply through transparent factor exposure.

This is one reason many active funds struggle to beat the market after fees. Before costs, some managers may match or slightly exceed a benchmark. After management fees, trading costs, taxes, and style drift, the aggregate result often becomes much less impressive. The arithmetic is difficult to escape: before costs, active management as a whole is the market; after costs, active management in aggregate must underperform it. William Sharpe made this point clearly in *The Arithmetic of Active Management*.

Once risk is measured, return has to be judged against it. This is the purpose of risk-adjusted performance metrics such as:

- **Sharpe ratio**, which scales excess return by total volatility;
- **Information ratio**, which scales active return by tracking error;
- **Sortino ratio**, which focuses on downside variation;
- **Maximum drawdown**, which captures peak-to-trough loss;
- **Beta and alpha estimates**, which separate broad exposure from residual return.

A small example is useful here too. Suppose two strategies both earn **8%** annualized while the risk-free rate is **2%**:

- Strategy X has 10% volatility.
- Strategy Y has 20% volatility.

Then:

- Sharpe(X) = \((8 - 2) / 10 = 0.60\)
- Sharpe(Y) = \((8 - 2) / 20 = 0.30\)

The return is the same, but the quality of that return is not. Raw return alone hides that distinction.

The practical implication is straightforward. Risk is not a correction applied after performance is observed. It is part of the economic meaning of performance itself. A result has value only when the uncertainty, drawdown profile, and exposure structure required to obtain it are made explicit.

## 4. Diversification and portfolio construction

Diversification is one of the few ideas in investing that is both simple and profound. Its importance does not come from the claim that risk can be eliminated, but from the recognition that risk can be shaped. Once assets are combined, the outcome depends not only on the expected return and volatility of each component, but also on how those components move together.

The whole cannot be understood by inspecting the parts in isolation.

This matters because investment thinking often begins in the wrong place. It is natural to focus first on selecting the best asset, the strongest signal, or the highest expected return. But a portfolio built from individually attractive components can still be fragile if those components are highly correlated or if they all depend on the same underlying economic force. A portfolio can often be made more resilient not by finding a better single idea, but by combining exposures whose return paths differ in meaningful ways.

That is why correlation matters so much. The value of adding a new position does not depend only on whether that position is attractive by itself. It depends on what it contributes to the portfolio already in place. A volatile asset may still improve the total portfolio if it behaves differently from existing holdings. Conversely, an asset with a favorable standalone profile may add little if it merely duplicates exposures already present.

The contribution of an asset is therefore marginal and relational, not absolute.

From this follows a broader principle: allocation is at least as important as selection. The same set of assets can produce very different outcomes under different weighting schemes. Equal weights, market-cap weights, volatility-based weights, and risk-budgeting approaches all encode different judgments about how capital and risk should be distributed.

None of these choices is neutral.

A few simple examples make the point more clearly.

### Example 1: same assets, different weights

Assume a portfolio of two assets:

- **Global equities** expected return: 8%, volatility: 16%
- **Investment-grade bonds** expected return: 4%, volatility: 6%

Now compare three allocations:

- **80/20** → expected return about 7.2%
- **60/40** → expected return about 6.4%
- **40/60** → expected return about 5.6%

The asset menu did not change. Only the weights changed. Yet the expected return, volatility, drawdown sensitivity, and dependence on equity markets all changed materially. Allocation is therefore not a secondary detail. It is one of the main decisions.

### Example 2: equal weight versus cap weight

Suppose four stocks have market values of:

- Stock A: **40**
- Stock B: **30**
- Stock C: **20**
- Stock D: **10**

A market-cap portfolio weights them **40/30/20/10**. An equal-weight portfolio assigns **25/25/25/25**.

That simple shift has consequences:

- Equal weight increases exposure to smaller names.
- It usually increases turnover because the portfolio must be rebalanced.
- It may improve diversification by reducing concentration in the largest holdings.
- It also changes the factor profile, often introducing a size tilt.

Again, the asset set is the same, but the portfolio is not.

### Example 3: why rebalancing matters

Suppose a portfolio starts at **50% equities / 50% bonds** with **€100**:

- **€50** in equities
- **€50** in bonds

After one year:

- equities rise **20%** → **€60**
- bonds rise **2%** → **€51**

The portfolio is now:

- equities: **€60**
- bonds: **€51**
- total: **€111**

The weights have drifted to roughly:

- equities: **54.1%**
- bonds: **45.9%**

The portfolio has already changed character. It is now more equity-heavy than intended. If that drift continues for several years, a portfolio that began as balanced can become increasingly concentrated in the recent winners.

Rebalancing restores the definition of the portfolio through time.

To return to **50/50**, the investor would sell about **€4.5** of equities and buy **€4.5** of bonds. That action does three things at once:

- it controls concentration,
- it maintains the intended risk structure,
- and it enforces discipline by trimming recent winners and adding to relative laggards.

This is why rebalancing should not be treated as a minor operational detail. It is one of the core mechanisms of portfolio design.

Different rebalancing rules embody different philosophies:

- **Calendar rebalancing**: monthly, quarterly, yearly;
- **Threshold rebalancing**: rebalance only when weights drift beyond a set band;
- **Risk-based rebalancing**: rebalance when volatility or factor contribution moves materially;
- **Cash-flow rebalancing**: use new contributions or withdrawals to restore target weights with less turnover.

Each rule implies trade-offs between stability, responsiveness, turnover, and tax cost. There is no universal best choice. But there is also no neutral choice.

This is one reason allocation deserves at least as much attention as selection. In many portfolios, long-run results are shaped more by broad exposure decisions than by security-level selection. A strong idea inside a poorly structured portfolio may still produce mediocre overall results. By contrast, a simpler set of assets, combined sensibly, can often do better than a collection of individually appealing positions arranged carelessly.

Diversification does have limits, and any serious framework has to acknowledge them. Correlations are not fixed. They often rise under stress, especially when liquidity contracts and investors are forced to reduce risk at the same time. Assets that appear well diversified in ordinary periods may become much more alike in crises. This does not make diversification useless. It means diversification must be judged dynamically rather than assumed mechanically.

The practical conclusion is that portfolio construction should be treated as a first-order research problem. In many cases, the most durable improvement in results will come not from a more aggressive signal, but from a better combination of exposures, a cleaner distribution of risk, and a more coherent rebalancing rule.

## 5. Behavioral limits and the limits to arbitrage

Any market framework that treats prices as the output of pure rational optimization is incomplete. Markets are populated by investors who overreact, hesitate, anchor on recent experience, extrapolate trends too far, and confuse noise with information.

Behavioral finance matters because it helps explain how information is interpreted and how prices can move in ways that are systematic without being fully rational. Daniel Kahneman and Amos Tversky, along with later work by Richard Thaler and others, made this difficult to ignore.

Overconfidence is perhaps the most immediate example. Investors regularly overestimate the quality of their information and the precision of their forecasts. In practice, this leads to excessive trading, weak diversification, and an exaggerated belief in one’s ability to detect turning points. The problem is not only psychological. It has portfolio consequences. A process built on overconfidence tends to trade too often, concentrate too early, and mistake conviction for evidence.

Herding is equally important. Investors do not process information independently. They react to prices, narratives, flows, and the behavior of other investors. This can produce self-reinforcing moves in which rising prices are taken as confirmation rather than as information already reflected in the market. Herding does not require panic in an extreme sense. It can emerge from benchmarking pressure, institutional incentives, career risk, and the simple fact that being wrong with everyone else is often safer professionally than being wrong alone.

Loss aversion adds another layer. In prospect theory, losses are usually experienced more intensely than gains of the same size. This distorts selling decisions, risk-taking, and portfolio adjustments. Investors may hold losing positions too long, realize gains too early, or reduce risk after a drawdown at precisely the wrong moment. That helps explain why realized portfolio outcomes often differ sharply from the outcomes implied by a more disciplined decision process.

These behavioral limits do not overturn the broader market view developed earlier. They do not imply that markets are easy to beat or that anomalies are available for easy extraction. In many cases, the opposite is true. Even when a behavioral distortion exists, it may be difficult to monetize because the timing is unstable, the path to convergence is uncertain, and interim losses may be severe.

This is where the limits to arbitrage enter. Mispricing can persist because exploiting it is risky, capital is finite, investors face redemption pressure, and the market can remain misaligned for longer than a strategy can tolerate. That observation remains important because it explains why recognizable distortions do not automatically translate into usable opportunity.

The practical value of behavioral finance therefore lies less in the idea of easy alpha and more in the design of robust process. Rules, diversification, and rebalancing have value not because they predict markets perfectly, but because they reduce the damage caused by impulsive decisions, trend-chasing, panic selling, and unjustified confidence.

One of the strongest arguments for a structured investment process is not that it eliminates uncertainty. It is that it limits the effect of predictable human error.

## 6. Rules-based portfolio design: factors, smart beta, and risk parity

Once the portfolio is treated as the true unit of analysis, the next question is how exposures should be weighted. Market-capitalization weighting is the default answer in public markets because it is simple, scalable, transparent, and low cost. But it is not the only possible answer, and it is not neutral. It embeds a particular distribution of capital across securities and sectors, and that distribution changes mechanically with price.

A rules-based portfolio design begins from the idea that the weighting scheme itself is a research choice.

Factor investing is the most direct expression of this logic. Instead of accepting market-cap weights as the default, factor-based designs tilt the portfolio toward characteristics associated with persistent differences in return behavior, such as value, size, momentum, profitability, quality, or lower volatility. The right way to think about these tilts is not as automatic sources of alpha, but as systematic departures from the market portfolio that may earn a premium, diversify existing exposures, or alter risk in a predictable way.

Their value has to be judged on exactly those terms:

- what exposure is being added,
- what benchmark is being left,
- what costs are being introduced,
- and whether the resulting portfolio is still attractive after turnover and implementation frictions.

Smart beta is best understood as the practical packaging of this idea. It refers to relatively passive, rules-based strategies designed to depart from conventional cap-weighting in the hope of improving risk-adjusted returns, reducing concentration, or gaining targeted factor exposure. Some approaches use value measures, some emphasize lower volatility or lower beta, some use equal weighting or fundamental weighting, and some blend several factors.

The attraction is clear: these strategies aim to preserve the transparency and discipline of indexing while changing the composition of the portfolio deliberately.

The caution is equally clear: once a rule departs from the market portfolio, it also departs from the market’s risk structure. That can improve results in some environments and weaken them in others.

A practical comparison helps. Suppose three equity funds produce the following long-run annualized returns:

- **Index fund**: **9.5%**
- **Smart beta value fund**: **10.1%**
- **Active stock-picking fund**: **10.0%**

On raw return alone, both non-index options look better. But now add frictions and exposures:

| Fund | Return | Fee | Likely tilt | Net interpretation |
|---|---:|---:|---|---|
| Index fund | 9.5% | 0.05% | Market beta | Cheap market exposure |
| Smart beta value fund | 10.1% | 0.35% | Value + slight size | Mild factor tilt, modest excess before fee |
| Active fund | 10.0% | 1.20% | Unclear; drifting | May lag net of fee unless true alpha exists |

After fees:

- Index fund net ≈ **9.45%**
- Smart beta fund net ≈ **9.75%**
- Active fund net ≈ **8.80%**

The active fund may have generated respectable gross performance and still failed to beat the index once costs are included. Meanwhile, the smart beta fund’s excess may be largely explained by transparent value exposure rather than manager skill. This does not make the strategy useless. It clarifies what the investor is actually buying.

Risk parity pushes the weighting problem one step further by shifting attention from capital weights to risk contributions. A portfolio can be dominated by one asset class even when the capital allocation appears balanced, because low-volatility and high-volatility assets contribute very differently to total portfolio risk.

For example, a **60/40** equity-bond portfolio often sounds balanced in capital terms, but in risk terms equities may contribute the majority of total volatility. A risk-parity approach attempts to scale exposures so that each sleeve contributes more evenly to total risk. The attraction is not that it avoids risk, but that it makes risk budgeting explicit.

Its weakness is also important. Once leverage is introduced to increase the contribution of low-volatility assets, the process becomes more sensitive to funding conditions, correlation shifts, and assumptions about future diversification. Risk parity is therefore not a universal solution. It is a particular answer to the question of how risk should be distributed across a portfolio.

The broader lesson of rules-based design is that every weighting scheme is a trade-off. A process may reduce concentration and improve diversification while increasing turnover. It may improve the balance of risk while introducing leverage. It may outperform in one environment and disappoint in another because its implicit exposures differ from those of the benchmark.

The practical issue is cost. Many managed strategies present themselves as improvements on simple passive exposure, but the improvement has to be judged after fees, turnover, taxes, and implementation friction. A more elaborate weighting rule is not valuable merely because it is more elaborate. If it mainly repackages market, value, size, momentum, or low-volatility exposure in a more expensive form, then the investor is not buying superior design so much as paying more for familiar risk. The relevant question is whether the strategy delivers a sufficiently better net portfolio to justify its added complexity and cost.

## 7. Taxes, costs, and compounding

It is easy to focus on gross return. It is more useful to ask how much of that return an investor actually keeps.

That is why taxes belong inside the framework of portfolio research rather than outside it. A strategy may look attractive before friction and much less attractive once turnover, realized gains, dividend taxation, withholding, and account structure are taken into account.

The central issue is not only that taxes reduce return. It is that they interrupt compounding.

A simple example shows why. Suppose two strategies both earn a gross **8%** annual return over many years:

- **Strategy A** is tax-efficient and loses only **0.5%** per year to taxes and costs.
- **Strategy B** turns over more and loses **2.0%** per year to taxes and costs.

Their net returns become:

- Strategy A: **7.5%**
- Strategy B: **6.0%**

Now start with **€100,000** for 20 years:

- At **7.5%**, final value ≈ **€425,000**
- At **6.0%**, final value ≈ **€321,000**

That difference of roughly **€104,000** is not a minor detail. It is the result of compounding on capital that Strategy B did not keep invested.

This is why turnover is a tax variable as much as a trading variable. A process that rebalances aggressively, rotates frequently, or realizes gains systematically must clear a higher bar than a slower-moving strategy. That does not mean high-turnover strategies are invalid. It means their apparent edge must be strong enough to survive not only transaction costs, but also the tax cost of realization.

Tax treatment also depends on structure. The same strategy can produce very different realized outcomes depending on where it is implemented:

- taxable account,
- tax-deferred account,
- tax-exempt vehicle,
- fund wrapper,
- derivative structure.

Dividends, coupon income, short-term gains, long-term gains, and derivative exposures may all be treated differently. A strategy cannot therefore be fully evaluated without specifying the tax context in which it is meant to operate. A process that is acceptable in a tax-deferred setting may be materially less attractive in a taxable one.

The practical implications are straightforward.

Tax-aware investing usually favors:

- lower unnecessary turnover,
- longer holding periods where feasible,
- selective realization of gains and losses,
- thoughtful asset location,
- and a clear distinction between strategic rebalancing and avoidable trading.

This does not imply that tax minimization should dominate every other objective. Sometimes a better portfolio is worth a taxable adjustment. But it does mean that taxes should be treated as part of the mechanism, not as an external inconvenience added at the end.

The same logic applies to costs more broadly. Fees, spreads, market impact, custody costs, financing costs, and operational friction all reduce the return that actually compounds. Small recurring costs can matter greatly over time.

A simple fee example makes the point. Suppose two funds both earn a gross **7%** annual return:

- Fund 1 charges **0.10%**
- Fund 2 charges **1.50%**

Net returns:

- Fund 1: **6.90%**
- Fund 2: **5.50%**

Over 25 years on **€100,000**:

- at **6.90%** → about **€529,000**
- at **5.50%** → about **€381,000**

The annual fee gap is only **1.40%**, but the long-run wealth gap is substantial. Compounding magnifies what remains invested and penalizes what is paid away.

For the purposes of this repository, the implication is simple. Any serious strategy evaluation should distinguish between:

- **gross return**,
- **net return before tax**,
- and **investor return after tax**, where relevant.

It should also state:

- expected turnover,
- likely realization pattern,
- account context,
- major implementation costs,
- and whether the strategy depends on assumptions that materially affect its attractiveness.

A framework that ignores taxes and costs may still produce interesting backtests. It does not yet describe the result that a real investor experiences.

## 8. Research protocol

A theoretical framework is useful only if it governs how future research is done. Without a protocol, even good principles remain too abstract to matter. The purpose of this section is therefore to define the minimum structure that any strategy study in this repository should follow.

The objective is not to impose a rigid template for its own sake. It is to ensure that each idea is examined in a way that makes its logic, evidence, and limitations visible.

A useful practical benchmark for this protocol is the methodology behind the SPIVA scorecards. SPIVA is not a substitute for original research, but it is a reminder that credible investment evaluation depends on category-appropriate benchmarking, explicit survivorship treatment, style-consistent classification, and long-horizon comparison rather than selective period choice. The point is not to declare active management impossible. It is to raise the evidentiary bar for any claim of skill.

### 1. State a clear hypothesis

Every study should begin with a hypothesis that is narrow enough to be testable and concrete enough to be falsifiable.

Good:
- “A 12-month momentum strategy in developed equities, rebalanced monthly, will outperform a cap-weight benchmark on a risk-adjusted basis.”

Weak:
- “Momentum seems promising.”
- “The market may be inefficient here.”

A useful hypothesis should specify:

- what is expected,
- under what conditions,
- over what horizon,
- and through what mechanism.

### 2. Give an economic rationale before the backtest

A process does not need an elaborate model, but it should have a reason for existing beyond historical fit. That reason may come from:

- risk compensation,
- behavioral bias,
- structural friction,
- valuation logic,
- or portfolio improvement.

Examples:

- “Value may work because investors overreact to poor recent fundamentals and because distressed assets carry risk that not all investors can bear.”
- “Trend following may work because information diffuses gradually and because investors adjust slowly to regime change.”

The rationale does not prove the strategy. It disciplines the search.

### 3. Make the data transparent

The universe, frequency, sample window, instrument definitions, rebalancing assumptions, and missing-data treatment should all be explicit. A strategy should be reproducible from its description.

Minimum checklist:

- asset universe,
- sample dates,
- data frequency,
- survivorship treatment,
- corporate-action treatment,
- signal definition,
- position-sizing rule,
- rebalance frequency,
- missing-data handling,
- transaction-cost assumptions.

Example:
- “US large-cap equities, monthly data, January 1995 to December 2025, top decile momentum signal, equal-weighted, monthly rebalanced, 20 bps one-way cost.”

If these elements remain vague, confidence in the result should fall.

### 4. Define the benchmark

No strategy should be evaluated in isolation. Each study should state the benchmark against which the process is being judged and explain why that benchmark is relevant.

Examples:

- For a US equity strategy: **S&P 500** or a broad US market index.
- For a long-short value strategy: a market-neutral baseline or a factor-matched alternative.
- For a tactical allocation model: a static 60/40 portfolio.
- For a currency-hedged strategy: the same portfolio unhedged.

The principle is always the same: the value of the strategy has to be measured relative to what could have been done more simply.

A useful minimum question is whether the result would still look persuasive under a SPIVA-style challenge: compared with an appropriate benchmark, with survivorship handled explicitly, with category or style consistency preserved, and over a long enough horizon to reduce the role of selective period choice.

### 5. Use a consistent performance framework

At a minimum, each study should report:

- annualized return,
- annualized volatility,
- maximum drawdown,
- Sharpe ratio,
- benchmark-relative return,
- and turnover.

Where relevant, it should also report:

- alpha and beta,
- information ratio,
- tracking error,
- downside deviation,
- hit rate,
- factor exposures,
- recovery time after drawdowns.

Example:
A strategy that earns **11%** versus a benchmark at **9%** may look strong. But if it does so with **22%** volatility, a **-45%** drawdown, and a beta of **1.4**, the interpretation changes. The extra return may be little more than a leveraged version of the same market exposure.

The purpose of the metrics is not to enlarge the table. It is to prevent raw return from dominating interpretation.

### 6. Test robustness

A result should be examined under reasonable changes in parameters, sample periods, and implementation assumptions.

Examples:

- If a momentum strategy works only with a **127-day** lookback but not with **6-, 9-, or 12-month** variants, confidence should fall.
- If a signal works only when rebalanced on the last trading day of the month but not a few days earlier or later, confidence should fall.
- If a strategy is strong from **2009–2021** but weak outside that window, the result may be regime-dependent rather than durable.

Robustness does not require perfect stability. It requires that the broad shape of the result survives reasonable perturbation.

### 7. Include implementation realism

Costs, turnover, liquidity, financing, borrow constraints, and execution assumptions should be treated as part of the strategy, not as an appendix.

Questions that should be answered include:

- Can the assets actually be traded at the assumed size?
- Are the rebalancing assumptions realistic?
- Are spreads and slippage included?
- Is shorting feasible where required?
- Is leverage available on the assumed terms?
- Does the strategy survive after these frictions?

Whenever the gross edge is small, this point becomes decisive.

### 8. State the failure conditions

Every study should identify what would weaken confidence in the result, what environments are likely to challenge it, and what assumptions it depends on.

Examples:

- “This strategy likely weakens in low-dispersion markets.”
- “This signal depends on stable trend persistence and may fail during sharp reversals.”
- “The result is sensitive to transaction costs above 40 bps.”
- “The diversification benefit disappears if stock-bond correlation turns persistently positive.”

A process that cannot specify its own failure conditions is much harder to evaluate honestly.

### 9. Separate insight from packaging

Many strategies look original because they are packaged originally. That is not the same thing as being economically distinct.

Every study should ask:

- Is this truly new information?
- Or is it market beta?
- Is this alpha?
- Or is it value, momentum, carry, duration, or small-cap exposure presented in a different form?

This step is especially important for funds and manager evaluation. A manager may appear skillful on raw return while merely harvesting common exposures in a more expensive format.

### 10. End with a plain-language conclusion

Every note should close by answering a small set of questions directly:

- What is the hypothesis?
- Why should it work?
- What data were used?
- What is the benchmark?
- How is success measured?
- Does the result survive variation and frictions?
- Under what conditions should confidence decline?

A research process that can answer those questions clearly is more valuable than one that produces stronger-looking backtests without a structure behind them.

### Minimal study template

For practical use, each future strategy note in this repository should be able to fit into the following skeleton:

1. **Hypothesis**
2. **Economic rationale**
3. **Data and universe**
4. **Portfolio construction and rebalancing**
5. **Benchmark**
6. **Performance summary**
7. **Risk and factor decomposition**
8. **Robustness checks**
9. **Costs, taxes, and implementation**
10. **Failure conditions**
11. **Conclusion**

That level of structure is helpful because it forces the argument to be explicit before conclusions are drawn.
